**Dataset: Preços de Combustíveis**

Referência: Agência Nacional do Petróleo, Gás Natural e Biocombustíveis (ANP) - Brasil: http://www.anp.gov.br/dados-estatisticos


**Justificativa da escolha:**
Os preços dos combustíveis têm um impacto direto nas finanças das pessoas e das empresas. São um componente importante dos custos operacionais para indústrias de transporte, logística, aviação, agricultura, entre outros. Além disso, os preços dos combustíveis também afetam o orçamento das famílias, especialmente aquelas que dependem de veículos para locomoção diária.

**Como pode ser útil:**

**Análise de Custos Operacionais:** Empresas que dependem de transporte de mercadorias podem usar os dados para analisar e prever os custos operacionais associados aos combustíveis. Isso pode ajudá-las a otimizar rotas, ajustar preços de produtos e tomar decisões de investimento.

**Gestão de Frota:** Empresas de transporte e logística podem usar os dados para monitorar e gerenciar o desempenho de sua frota, identificar padrões de consumo de combustível e implementar estratégias para melhorar a eficiência do combustível.

**Planejamento Financeiro Pessoal:** Indivíduos podem usar os dados para acompanhar os preços dos combustíveis e planejar seus gastos com combustível. Eles podem ajustar seus orçamentos e hábitos de condução com base nas flutuações de preço.

**Análise Econômica:** Economistas e analistas financeiros podem usar os dados para entender as tendências de preços dos combustíveis e seu impacto na economia em geral. Isso pode incluir análises de inflação, custo de vida e impacto nos setores de transporte e logística.

**Políticas Públicas:** Governos podem usar os dados para monitorar os preços dos combustíveis e implementar políticas que visem estabilizar os preços, garantir a segurança energética e promover fontes de energia alternativas.

**Análise de Mercado:** Empresas do setor de combustíveis e varejistas podem usar os dados para monitorar a concorrência, ajustar suas estratégias de preços e tomar decisões de marketing.

**Conclusão:**
Os preços dos combustíveis são uma variável financeira importante que afeta diversos setores da economia e o orçamento das famílias. Portanto, um conjunto de dados sobre preços de combustíveis pode ser uma fonte valiosa de informações para análises financeiras, planejamento estratégico e tomada de decisões tanto no setor privado quanto no público.

**Regiao:**

**Tipo de dado: Categórico (Texto)**
Descrição: Indica a região geográfica do Brasil onde o posto de gasolina está localizado.
Observações: Esta variável inclui seis regiões: Norte (N), Nordeste (NE), Centro-Oeste (CO), Sudeste (SE), Sul (S).

**Estado:**

**Tipo de dado: Categórico (Texto)**
Descrição: Indica o estado do Brasil onde o posto de gasolina está localizado.
Observações: Esta variável indica o estado específico dentro da região.
Municipio:

**Tipo de dado: Categórico (Texto)**
Descrição: Indica o município onde o posto de gasolina está localizado.

**Revenda:**

**Tipo de dado: Categórico (Texto)**
Descrição: Nome da revenda de combustível.
Observações: Esta variável identifica o nome da empresa que opera o posto de gasolina.

**CNPJ da Revenda:**

**Tipo de dado: Categórico (Texto)**
Descrição: Número de Cadastro Nacional da Pessoa Jurídica (CNPJ) da revenda de combustível.
Observações: Identifica exclusivamente cada revenda de combustível.

**Nome da Rua:**

**Tipo de dado: Categórico (Texto)**
Descrição: Nome da rua onde o posto de gasolina está localizado.

**Numero Rua:**

**Tipo de dado: Categórico (Texto)**
Descrição: Número do endereço do posto de gasolina.

**Complemento:**

**Tipo de dado: Categórico (Texto)**
Descrição: Informações adicionais sobre o endereço do posto de gasolina.

**Bairro:**

**Tipo de dado: Categórico (Texto)**
Descrição: Nome do bairro onde o posto de gasolina está localizado.

**Cep:**

**Tipo de dado: Categórico (Texto)**
Descrição: Código de Endereçamento Postal (CEP) do posto de gasolina.

**Produto:**

**Tipo de dado: Categórico (Texto)**
Descrição: Tipo de produto de combustível vendido pelo posto.
Observações: Esta variável inclui diferentes tipos de produtos de combustível, como Gasolina, Diesel, Álcool, etc.

**Data da Coleta:**

**Tipo de dado: Data**
Descrição: Data em que os dados foram coletados.
Observações: Esta variável indica quando os preços dos combustíveis foram registrados.

**Valor de Venda:**

**Tipo de dado: Numérico (Float)**
Descrição: Preço de venda do combustível.
Observações: Esta variável representa o preço pelo qual o combustível está sendo vendido no posto de gasolina.

**Valor de Compra:**

**Tipo de dado: Numérico (Float)**
Descrição: Preço de compra do combustível.
Observações: Esta variável parece ter apenas um valor e pode não ser útil para análise.

**Unidade de Medida:**

**Tipo de dado: Categórico (Texto)**
Descrição: Unidade de medida para o preço de venda e compra do combustível.
Observações: Geralmente representa 'R$/litro', indicando o preço por litro do combustível.

**Bandeira:**

**Tipo de dado: Categórico (Texto)**
Descrição: Marca ou bandeira associada ao posto de gasolina.
Observações: Identifica a marca ou a empresa associada ao posto de gasolina.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/BaronOfBattle/bigdatamaladrage/main/combustiveis.csv"
df = pd.read_csv(url, sep=';', names=['Regiao', 'Estado', 'Municipio', 'Revenda', 'CNPJ da Revenda', 'Nome da Rua', 'Numero Rua', 'Complemento', 'Bairro', 'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda', 'Valor de Compra', 'Unidade de Medida', 'Bandeira'])

df['Valor de Venda'] = pd.to_numeric(df['Valor de Venda'], errors='coerce')

# remove a coluna 'Valor de Compra' (apenas 1 valor)
df.drop(columns=['Valor de Compra'], inplace=True)

# verifica os tipos de dados de cada coluna
tipos_de_dados = df.dtypes
print("Tipos de dados de cada coluna:")
print(tipos_de_dados)
print("\n")

# resumo estatístico das variáveis numéricas
resumo_estatistico = df.describe()
print("Resumo estatístico das variáveis numéricas:")
print(resumo_estatistico)
print("\n")

# visualização da distribuição do valor de venda
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Valor de Venda', bins=30, kde=True)
plt.title('Distribuição do Valor de Venda')
plt.xlabel('Valor de Venda')
plt.ylabel('Frequência')
plt.show()

# visualização do valor de venda por produto
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Produto', y='Valor de Venda')
plt.title('Distribuição do Valor de Venda por Produto')
plt.xlabel('Produto')
plt.ylabel('Valor de Venda')
plt.show()
```

Gráfico 1: Distribuição do Valor de Venda

O gráfico apresenta a distribuição dos valores de venda dos produtos de combustível.

O eixo x representa os valores de venda, indicando o preço pelo qual os produtos foram vendidos.

O eixo y representa a frequência, ou seja, o número de ocorrências de cada valor de venda.

As barras verticais no gráfico representam a contagem de ocorrências dos valores de venda em cada intervalo (bin). Neste caso, foram utilizados 30 bins para agrupar os valores de venda.

A linha suave que atravessa o histograma é a estimativa da densidade kernel (KDE), que suaviza a distribuição dos dados e ajuda a visualizar a forma da distribuição.

Gráfico 2: Distribuição do Valor de Venda por Produto



O gráfico apresenta a distribuição dos preços de venda dos produtos de combustível por produto.

Os preços de venda são representados no eixo y, enquanto os diferentes produtos de combustível são representados no eixo x.

Cada boxplot mostra a distribuição dos preços de venda para um produto específico.

Dentro de cada boxplot, a linha no meio da caixa representa a mediana dos preços de venda para o produto correspondente.

A caixa abrange o intervalo interquartil (IQR), que contém a maioria dos dados, indo do primeiro quartil (25%) ao terceiro quartil (75%).

Pontos individuais além das "whiskers" são considerados outliers e são plotados separadamente.

```python

url = "https://raw.githubusercontent.com/BaronOfBattle/bigdatamaladrage/main/combustiveis.csv"
df = pd.read_csv(url, sep=';', names=['Regiao', 'Estado', 'Municipio', 'Revenda', 'CNPJ da Revenda', 'Nome da Rua', 'Numero Rua', 'Complemento', 'Bairro', 'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda', 'Valor de Compra', 'Unidade de Medida', 'Bandeira'])

# converter coluna 'Valor de Venda' para numérica, lidando com valores não convertíveis
df['Valor de Venda'] = pd.to_numeric(df['Valor de Venda'], errors='coerce')

# remove a coluna 'Valor de Compra' (apenas 1 valor)
df.drop(columns=['Valor de Compra'], inplace=True)

# seleciona apenas os 20 valores mais frequentes para as variáveis categóricas com muitos valores únicos (sem isso fica horroroso os graficos)
top_20_values = ['Cep', 'Bairro', 'Complemento', 'Municipio', 'Revenda', 'CNPJ da Revenda', 'Nome da Rua', 'Numero Rua']
for col in top_20_values:
    top_20 = df[col].value_counts().nlargest(20).index
    df.loc[~df[col].isin(top_20), col] = 'Outros'

# gráficos descritivos para cada variável
for col in df.columns:
    plt.figure(figsize=(10, 6))
    if df[col].dtype == 'object':
        sns.countplot(data=df, x=col)
        plt.title(f'Distribuição de {col}')
        plt.xticks(rotation=45, ha='right')
    else:
        sns.histplot(data=df, x=col, bins=30, kde=True)
        plt.title(f'Distribuição de {col}')
    plt.xlabel(col)
    plt.ylabel('Frequência')
    plt.show()
```

#Comparação das Estatísticas Descritivas dos Preços de Venda por Produto


```python

from scipy.stats import chi2_contingency


# seleciona apenas as colunas categóricas
categorical_vars = df.select_dtypes(include='object')

# Calcula tabelas de contingência e realiza o teste qui-quadrado
for col1 in categorical_vars.columns:
    for col2 in categorical_vars.columns:
        if col1 != col2:
            contingency_table = pd.crosstab(df[col1], df[col2])
            print(f'Tabela de Contingência para {col1} e {col2}:')
            print(contingency_table)
            print("\n")

            # Realiza o teste qui-quadrado para avaliar a associação entre as variáveis
            chi2, p, _, _ = chi2_contingency(contingency_table)
            print(f"Teste qui-quadrado para {col1} e {col2}:")
            print(f"Qui-quadrado: {chi2}")
            print(f"Valor p: {p}")
            print("\n")
```
